# Extraigo información de todas las paginas al mismo tiempo

```python
tabla_pag_1= pdf[indices_detalle[0]:-1]
tabla_pag_1
```

```python
#fechas
fecha_lista=[]
for i in range(len(tabla_pag_1)):
    if re.match('fecha',tabla_pag_1[i].lower()):
        fecha_lista.append(tabla_pag_1[i+1])
        
fecha_lista
```

```python
#separo
fecha_lista_2=[]
for i in fecha_lista:
    j=re.findall(r'\d{2} \w{3}',i)
    fecha_lista_2.append(j)
```

```python
#aplano
fecha_lista_3=[]
for sublist in fecha_lista_2:
    for item in sublist:
        fecha_lista_3.append(item)
```

```python
print(fecha_lista_3)
```

```python
#conceptos
conceptos_lista_pag_1=[]
for i in range(len(tabla_pag_1)):
    if re.match('concepto',tabla_pag_1[i].lower()):
        conceptos_lista_pag_1.append(tabla_pag_1[i+1])
        conceptos_lista_pag_1.append(tabla_pag_1[i+2])
        
conceptos_lista_pag_1
```

```python
conceptos_lista_2=[]
for i in conceptos_lista_pag_1:
    conceptos_prueba1=re.split('SALDO | INTERESES |INTERESES | RETIRO | DEP',i)
    conceptos_lista_2.append(conceptos_prueba1)
```

```python
#aplano
conceptos_lista_3=[]
for sublist in conceptos_lista_2:
    for item in sublist:
        conceptos_lista_3.append(item)
conceptos_lista_3
```

```python
conceptos_pag_1=[]
for i in range(len(conceptos_lista_3)):
    if conceptos_lista_3[i].startswith('ANTERIOR'):
        j='SALDO '+conceptos_lista_3[i]
        conceptos_pag_1.append(j)
    elif conceptos_lista_3[i].startswith('AL'):
        j='INTERES '+conceptos_lista_3[i]
        conceptos_pag_1.append(j)
    elif conceptos_lista_3[i].isdigit():
        j='RETIRO '+ conceptos_lista_3[i]
        conceptos_pag_1.append(j)
    elif conceptos_lista_3[i].startswith('OSITO'):
        j='DEP'+ conceptos_lista_3[i]
        conceptos_pag_1.append(j)

conceptos_pag_1
```

```python
#retiros
retiros_lista_pag_1=[]
for i in range(len(tabla_pag_1)):
    if re.match('retiros',tabla_pag_1[i].lower()):
        retiros_lista_pag_1.append(tabla_pag_1[i])
        retiros_lista_pag_1.append(tabla_pag_1[i+1])
        retiros_lista_pag_1.append(tabla_pag_1[i+2])
        retiros_lista_pag_1.append(tabla_pag_1[i+3])
retiros_lista_pag_1
```

```python
retiros1_0=[]
for i in retiros_lista_pag_1:
    j=i.split('RETI')
    retiros1_0.append(j)
retiros1_0
```

```python
#aplano
retiros1_1=[]
for sublist in retiros1_0:
    for item in sublist:
        retiros1_1.append(item)
retiros1_1        
```

```python
retiros1_2=[]
for i in range(len(retiros1_1)):
    if retiros1_1[i].startswith('ROS'):
        retiros1_2.append(retiros1_1[i])
        if re.search(r'\d+',retiros1_1[i+1]):
            retiros1_2.append(retiros1_1[i+1])
            if re.search(r'\d+',retiros1_1[i+2]):
                retiros1_2.append(retiros1_1[i+2])
                if re.search(r'\d+',retiros1_1[i+3]):
                    retiros1_2.append(retiros1_1[i+3])
                    
                
        
```

```python
retiros1_3=[]
for i in retiros1_2:
    if i.startswith('DEPOSITO'):
        pass
    else:
        retiros1_3.append(i)
```

```python
retiros1_4=[]
for i in retiros1_3:
    j=i.split(' ')
    retiros1_4.append(j)
```

```python
#aplano
retiros1_5=[]
for sublist in retiros1_4:
    for item in sublist:
        if re.search(r'\d+',item):
            retiros1_5.append(item)
```

```python
#organizo retiros todas las paginas
retiros_pag_1=[]
contador=0
for i in range(len(conceptos_pag_1)):
    if conceptos_pag_1[i].startswith('RETIRO'):
        retiros_pag_1.append(retiros1_5[contador])
        contador=contador+1
    else:
        retiros_pag_1.append('0')
retiros_pag_1
```

```python
#depositos
depositos_lista_pag_1=[]
for i in range(len(tabla_pag_1)):
    if re.match('depositos',tabla_pag_1[i].lower()):
        depositos_lista_pag_1.append(tabla_pag_1[i])
        depositos_lista_pag_1.append(tabla_pag_1[i+1])
        depositos_lista_pag_1.append(tabla_pag_1[i+2])
        depositos_lista_pag_1.append(tabla_pag_1[i+3])
        depositos_lista_pag_1.append(tabla_pag_1[i+4])
        depositos_lista_pag_1.append(tabla_pag_1[i+5])
        depositos_lista_pag_1.append(tabla_pag_1[i+6])
        
depositos_lista_pag_1
```

```python
depositos_0=(' '.join(depositos_lista_pag_1)).split('DEP')
```

```python
depositos_1=[]
for i in depositos_0:
    depositos_1.append(i.split('SALDO'))
```

```python
#aplano
deposito_2=[]
for sublist in depositos_1:
    for item in sublist:
        if item.startswith('OSITO'):
            deposito_2.append(item)

        
```

```python
deposito_3=[]
for i in deposito_2:
    j= i.split(' ')
    deposito_3.append(j)
```

```python
#aplano
deposito_4=[]
for sublist in deposito_3:
    for item in sublist:
        if re.search(r'\d+', item):
            deposito_4.append(item)
```

```python
deposito_4
```

```python
#organizo depositos
depositos_pag_1=[]
contador=0
for i in range(len(conceptos_pag_1)):
    if conceptos_pag_1[i].startswith('DEPOSITO') or conceptos_pag_1[i].startswith('INTERES'):
        depositos_pag_1.append(deposito_4[contador])
        contador=contador+1
    else:
        depositos_pag_1.append('0')
depositos_pag_1
```

```python
#saldos
saldos_lista_pag_1=[]
for i in range(len(tabla_pag_1)):
    if re.match('saldo',tabla_pag_1[i].lower()):
        saldos_lista_pag_1.append(tabla_pag_1[i])
        saldos_lista_pag_1.append(tabla_pag_1[i+1])
        saldos_lista_pag_1.append(tabla_pag_1[i+2])
        saldos_lista_pag_1.append(tabla_pag_1[i+3])
        saldos_lista_pag_1.append(tabla_pag_1[i+4])
        saldos_lista_pag_1.append(tabla_pag_1[i+5])
        saldos_lista_pag_1.append(tabla_pag_1[i+6])
        saldos_lista_pag_1.append(tabla_pag_1[i+7])
        
saldos_lista_pag_1
```

```python
saldos1_1=' '.join(saldos_lista_pag_1)
saldos1_1
```

```python
saldo1_2=[]
for i in saldos1_1.split('SAL'):
    if re.search(r'DO \d+',i):
        saldo1_2.append(i)
saldo1_2        
```

```python
saldo1_3=[]
for i in saldo1_2:
    j=re.findall(r'\d+\,\d+\.\d+',i)
    saldo1_3.append(j)
```

```python
#aplano
saldo1_4=[]
for sublist in saldo1_3:
    for item in sublist:
        saldo1_4.append(item)
```

```python
#organizo saldos
saldos_pag_1=[]
contador=0
for i in range(len(conceptos_pag_1)):
    if conceptos_pag_1[i].startswith('SALDO') or conceptos_pag_1[i].startswith('INTERES') :
        saldos_pag_1.append(saldo1_4[contador])
        contador=contador+1
    else:
        saldos_pag_1.append('0')
saldos_pag_1
```
