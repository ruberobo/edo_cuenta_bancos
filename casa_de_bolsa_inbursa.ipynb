```python
import pandas as pd
import numpy as np
import re
import textract
```

```python
import tabula
```

```python
#extraigo texto
archivo= 'tresalia\Bancos\inbursa\pdf\Casa_de_Bolsa_Inbursa.pdf'
```

```python
text = textract.process(archivo)
```

```python
#cambio formato
encoding = 'utf-8'
text_str=text.decode(encoding)
```

```python
#separo
pdf=re.split(r"[~\r\n]+", text_str)
```

```python
information_pdf=pd.DataFrame(pdf).rename(columns={0:"informacion"})
information_pdf.head()
```

```python
pdf
```

```python
contrato_lista=[]
for i in range(len(pdf)):
    if re.search('CONTRATO',pdf[i]):
        contrato_lista.append(pdf[i])
        contrato_lista.append(pdf[i+1])
contrato=contrato_lista[1]
contrato
```

```python
cliente_lista=[]
for i in pdf:
    if re.search('CLIENTE',i):
        cliente_lista.append(i)
cliente=cliente_lista[0]
cliente
```

```python
nombre_lista=[]
for i in pdf:
    if re.search('APODERADO',i):
        nombre_lista.append(i)
nombre=nombre_lista[0]
nombre
```

```python
periodo_lista=[]
for i in pdf:
    if re.search('PERIODO',i):
        periodo_lista.append(i)
periodo=periodo_lista[0]
periodo
```

```python
rfc_lista=[]
for i in pdf:
    if re.search('rfc',i) or re.search('RFC',i) or re.search('R.F.C.',i) or re.search('R.F. .C',i):
        rfc_lista.append(i)
rfc=rfc_lista[-1]
rfc
```

```python
#direccion
direccion_lista=[]
for i in pdf:
    if re.search('direccion',i) or re.search('DIRECCION',i) or re.search('DIRECCIÓN',i) or re.search('Direccion',i):
        direccion_lista.append(i)
direccion=direccion_lista[0]
direccion
```

```python
#moneda
moneda_lista=[]
for i in pdf:
    if re.search('moneda',i) or re.search('MONEDA',i) or re.search('Moneda',i):
        moneda_lista.append(i)
moneda=moneda_lista[0]
moneda
```

```python
tabla=tabula.read_pdf(archivo, pages='all')
```

```python
tabla[3]
```

```python
#traspuesta de la traspuesta para convertir el encabezado en una fila mas
tabla_mov=tabla[3].T.reset_index().T
```

```python
tabla_mov
```

```python
movi=pd.DataFrame()
movi['días op']=tabla_mov[0]
movi['días LQ']=tabla_mov[1]
movi['días *']= tabla_mov[2]
```

```python
movi['emisora']=tabla_mov[3]+tabla_mov[4]+tabla_mov[5]
```

```python
movi['número de títulos']=tabla_mov[8]
movi['precio unitario']=tabla_mov[9]
movi['comision']=tabla_mov[11]
movi['IVA']=tabla_mov[12]
movi['ISR']=tabla_mov[14]
movi['neto de operación']=tabla_mov[16]
movi['saldo efectivo']=tabla_mov[17]
```

```python
movi_2=movi[movi['emisora'].isnull()==False]
```

```python
movi_2['precio unitario']=movi_2['precio unitario'].replace('◄','4')
movi_2['neto de operación']=movi_2['neto de operación'].replace('◄','4')
```

```python
movi_2['precio unitario']=movi_2['precio unitario'].replace('o','0')
```

```python
movimientos=movi_2.fillna(0).reset_index(drop=True)
```
